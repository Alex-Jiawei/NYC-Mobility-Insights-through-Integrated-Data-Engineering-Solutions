## City Bike Data Preprocess

```python
import pandas as pd
import os
```

```python
directory_path = '/Volumes/Extreme SSD/DEP'
file_paths = [
    os.path.join(directory_path, 'combined_data_2022.csv'),
    os.path.join(directory_path, 'combined_data_2023.csv'),
    os.path.join(directory_path, 'combined_data_2024.csv')
]
```

```python
chunk_list = []

# Date range
start_date = "2022-03-01"
end_date = "2024-03-31"
```

```python
for file in file_paths:
    print(f"Processing {file} in chunks...")
    for chunk in pd.read_csv(file, chunksize=100000):
        
        chunk['started_at'] = pd.to_datetime(chunk['started_at'], errors='coerce', format='mixed')
        
        chunk = chunk.dropna(subset=['started_at'])

        chunk = chunk[(chunk['started_at'] >= start_date) & (chunk['started_at'] <= end_date)]
        
        chunk['time_id'] = chunk['started_at'].dt.strftime('%Y-%m-%d-%H')
        
        chunk.drop(columns=['rideable_type', 'member_casual'], inplace=True)
        
        chunk_list.append(chunk)
```

```python
combined_df = pd.concat(chunk_list, ignore_index=True)
print("All files processed, filtered, and combined successfully.")
```

```python
combined_df
```

```python
output_file = os.path.join(directory_path, "filtered_citi_bike_data_large.csv")
combined_df.to_csv(output_file, index=False)
print(f"Filtered data saved to {output_file}.")
```

#### citi_bike_stations Table

```python
import pandas as pd
```

```python
data_path = '/Volumes/Extreme SSD/DEP/filtered_citi_bike_data_large.csv'
```

```python
stations_start_list = []
stations_end_list = []

chunk_size = 100000
```

```python
print("Processing the large file in chunks...")
for chunk in pd.read_csv(data_path, chunksize=chunk_size):

    stations_start = chunk[['start_station_id', 'start_station_name', 'start_lat', 'start_lng']]
    stations_start.columns = ['station_id', 'station_name', 'latitude', 'longitude']
    stations_start_list.append(stations_start)
    

    stations_end = chunk[['end_station_id', 'end_station_name', 'end_lat', 'end_lng']]
    stations_end.columns = ['station_id', 'station_name', 'latitude', 'longitude']
    stations_end_list.append(stations_end)
```

```python
stations_start_combined = pd.concat(stations_start_list, ignore_index=True)
stations_end_combined = pd.concat(stations_end_list, ignore_index=True)

citi_bike_stations = pd.concat([stations_start_combined, stations_end_combined]).drop_duplicates().reset_index(drop=True)
```

```python
citi_bike_stations['latitude'] = pd.to_numeric(citi_bike_stations['latitude'], errors='coerce')
citi_bike_stations['longitude'] = pd.to_numeric(citi_bike_stations['longitude'], errors='coerce')

citi_bike_stations_tem = citi_bike_stations.dropna(subset=['latitude', 'longitude'])
```

```python
citi_bike_stations_tem['latitude'] = citi_bike_stations_tem['latitude'].round(3)
citi_bike_stations_tem['longitude'] = citi_bike_stations_tem['longitude'].round(3)
```

```python
citi_bike_stations_list = citi_bike_stations_tem.drop_duplicates(subset=['station_id']).reset_index(drop=True)
```

```python
citi_bike_stations_list
```

```python
citi_bike_stations_list['station_id'] = citi_bike_stations_list['station_id'].astype(str)

nyc_stations = citi_bike_stations_list[citi_bike_stations_list['station_id'].str.match(r'^\d{4}\.\d{2}$')]

nyc_stations['station_id'] = pd.to_numeric(nyc_stations['station_id'])

nyc_stations_sorted = nyc_stations.sort_values(by='station_id', ascending=True).reset_index(drop=True)
```

```python
nyc_stations_sorted
```

```python
# Drop duplicate rows based on `station_id`
citi_bike_stations_table = nyc_stations_sorted.drop_duplicates(subset=['station_id']).reset_index(drop=True)
```

```python
citi_bike_stations_table
```

```python
eda_summary = {
    "Total Stations": citi_bike_stations_table['station_id'].nunique(),
    "Total Unique Names": citi_bike_stations_table['station_name'].nunique(),
    "Latitude Range": (citi_bike_stations_table['latitude'].min(), citi_bike_stations_table['latitude'].max()),
    "Longitude Range": (citi_bike_stations_table['longitude'].min(), citi_bike_stations_table['longitude'].max()),
    "Most Common Station Name": citi_bike_stations_table['station_name'].mode()[0],
    "Most Common Latitude": citi_bike_stations_table['latitude'].mode()[0],
    "Most Common Longitude": citi_bike_stations_table['longitude'].mode()[0]
}

print(eda_summary)
```

```python
output_path = 'citi_bike_stations.csv'
citi_bike_stations_table.to_csv(output_path, index=False)
print(f"citi_bike_stations table saved to {output_path}")
```

```python
import pandas as pd

# File paths (update with your actual paths)
files = [
    "/Volumes/Extreme SSD/DEP/combined_data_2022.csv",
    "/Volumes/Extreme SSD/DEP/combined_data_2023.csv",
    "/Volumes/Extreme SSD/DEP/combined_data_2024.csv"
]

# Function to check rows and columns for large files
def get_file_shape(file_path):
    chunksize = 100000
    total_rows = 0
    for chunk in pd.read_csv(file_path, chunksize=chunksize):
        total_rows += len(chunk)
    # Get the number of columns from the first chunk
    columns = chunk.columns if 'chunk' in locals() else []
    return total_rows, len(columns)

# Iterate through files and collect their shapes
file_shapes = {}
for file in files:
    try:
        rows, columns = get_file_shape(file)
        file_shapes[file] = {"Rows": rows, "Columns": columns}
    except Exception as e:
        file_shapes[file] = {"Error": str(e)}

# Convert the results into a DataFrame and display
file_shapes_df = pd.DataFrame(file_shapes).T
file_shapes_df.columns = ["Rows", "Columns"]
print(file_shapes_df)
```
